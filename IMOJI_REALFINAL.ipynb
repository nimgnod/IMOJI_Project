```python
import webcolors 

from webcolors import name_to_rgb, rgb_to_name

from openpyxl import load_workbook

# 스크린샷에서 좌표값 구하는 부분 #

from PIL import ImageGrab

screen = ImageGrab.grab()

import pyautogui as pag

select = pag.position()

screen.getpixel(select) # getpixel(좌표값)

pick = screen.getpixel(select)

# 여기까지 #


# 색깔 리스트

Colors = {'gold':'골드',
          'gray':'그레이', 
          'green':'그린',
          'greenyellow':'네온',
          'navy':'네이비', 
          'lavender':'라벤더',
          'red':'레드',
          'cyan':'민트',
          'beige':'베이지',
          'black':'블랙',
          'blue':'블루',
          'skyblue':'스카이블루', 
          'silver':'실버',
          'yellow':'옐로우',
          'orange':'오렌지', 
          'firebrick':'와인',
          'khaki':'카키',
          'purple':'퍼플',
          'pink':'핑크',
          'white':'화이트'
         }

# 입력값 받는 부분

from math import sqrt

# 색깔 목록
COLORS = (
    name_to_rgb('gold'),
    name_to_rgb('gray'),
    name_to_rgb('green'),
    name_to_rgb('greenyellow'),
    name_to_rgb('navy'),
    name_to_rgb('lavender'),
    name_to_rgb('red'),
    name_to_rgb('cyan'),
    name_to_rgb('beige'),
    name_to_rgb('brown'),
    name_to_rgb('black'),
    name_to_rgb('blue'),
    name_to_rgb('skyblue'),
    name_to_rgb('silver'),
    name_to_rgb('yellow'),
    name_to_rgb('orange'),
    name_to_rgb('firebrick'),
    name_to_rgb('khaki'),
    name_to_rgb('purple'),
    name_to_rgb('pink'),
    name_to_rgb('white')
)

def closest_color(rgb):
    r, g, b = rgb
    color_diffs = []
    for color in COLORS:
        cr, cg, cb = color
        color_diff = sqrt(abs(r - cr)**2 + abs(g - cg)**2 + abs(b - cb)**2)
        color_diffs.append((color_diff, color))
    return min(color_diffs)[1]

load_wb = load_workbook("C:/IMOJI_DATASET.xlsx", data_only=True)

f_pick = rgb_to_name(closest_color(pick))

print('선택한 상의 색깔은 ' + Colors[f_pick] + ' 입니다')
print('선택한 상의 색깔에 많이 입은 하의 색깔 순위')

load_ws = load_wb['{}'.format(Colors[f_pick])]

for i in range(1, 22):
    if load_ws['F{}'.format(i+1)].value == 0:
        break
    print(load_ws['D{}'.format(i+1)].value,'순위 : ', load_ws['E{}'.format(i+1)].value)
```
